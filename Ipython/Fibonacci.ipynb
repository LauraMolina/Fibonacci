```python
import matplotlib.pyplot as plt
%matplotlib inline
from time import time

x = []
y = []
n = 100
m = 10000
p = 100
for n in range(n,m,p):
    a = 1
    b = 0
    time_i = time()
    for i in range(n):
        b = a - b
        a = a + b
    time_f = time()
    time_e = time_f - time_i
    x.append(n)
    y.append(time_e)
plt.plot(x, y)
plt.xlabel('Fibonnaci(n)')
plt.ylabel('Tiempo de ejecucion')
plt.title('Tiempo de ejecucion del algoritmo de Fibonnaci')
plt.show()
```
